```python
def get_genre(x):
    gen=[]
    for i in x:
        if i['name']=='Science Fiction':
            name='Sci-Fi'
            gen.append(name)
        else:
            gen.append(i['name'])
    if gen==[]:
        return []
    else:
        return ' '.join(gen)
```

```python
film=Movie()
def get_genre(x):
    genre=[]
    res=film.search(x)
    id=res[0].id
    response=requests.get('https://api.themoviedb.org/3/movie/{}?api_key=83cf022ec205972d97571d13de6d4b02'.format(id))
    data=response.json()
    for i in range(0,len(data['genres'])):
          genre.append(data['genres'][i]['name'])      
    return ' '.join(genre)

def get_keyword(x):
    key=[]
    res=film.search(x)
    key=list(film.keywords(res[0].id))
    return ' '.join(str(i) for i in key)
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (director)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
    
```

```python
def actor1(x):
    return (x.split("screenplay); ")[-1].split(", ")[0])

def actor2(x):
    if len(x.split("screenplay); ")[-1].split(", "))<2:
        return np.nan
    else:
        return (x.split("screenplay); ")[-1].split(", ")[1])
    
def actor3(x):
    if len(x.split("screenplay); ")[-1].split(", "))<3:
        return np.nan
    else:
        return (x.split("screenplay); ")[-1].split(", ")[2])
     
```

```python
def prepare(df):
    df['genres']=df['Title'].apply(get_genre)
    df['keywords']=df['Title'].apply(get_keyword)
    df['director_name']=df['Cast and crew'].apply(get_director)
    df['actor_1_name']=df['Cast and crew'].apply(actor1)
    df['actor_2_name']=df['Cast and crew'].apply(actor2)
    df['actor_3_name']=df['Cast and crew'].apply(actor3)
    df.drop(['Opening','Opening.1','Production company','Cast and crew','Ref.'],axis=1,inplace=True)
    df.rename(columns={'Title':'movie_title'},inplace=True)
    return df
```

```python
def clean_data(x):
    if isinstance(x,list):
        return [str.lower(i.replace(" ",'')) for i in x ]
    else:
        if isinstance(x,str):
            return str.lower(x.replace(' ',''))
        else:
            return ''
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn import metrics
vect=TfidfVectorizer(stop_words='english',analyzer='word',ngram_range=(1,1),use_idf=True)
matrix=vect.fit_transform(old_data['soup'])
```

```python
cosine_sim=metrics.pairwise.linear_kernel(matrix,matrix)
```

```python
def recommend(movie,number,sim=cosine_sim):
    idx=moviesid[movie]
    li=list(enumerate(sim[idx]))
    li=sorted(li,key = lambda x:x[1] , reverse=True)
    li=li[1:number+1]
    res=[x[0] for x in li]
    return old_data['movie_title'].iloc[res]
```

```python
recommend('Inception',15)
```
